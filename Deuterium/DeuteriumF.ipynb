By applying the continuity of the wavefunction and that of the derivative of the wavefunction and then dividing by parts, we obtain:
$$ kL cot(kL) = - \gamma L$$
This function provides us the possible energy values of the bounded states. We will solve this equation graphically by firstly setting $ \xi = kL \geq 0$ and $ \nu = \gamma L \geq 0$. Furthermore, we use some algebra in order to obtain the following equations:  
$$ \nu = - \xi cot( \xi)$$
$$ \nu = \sqrt{\frac{2mV_0}{\hbar^2}L^2 - \xi^2}$$

The section of points of those two function give the possible values of $\xi = \xi (E)$. Because we need to find the solutions with only ONE bounded state, the existance of the cot function gives the following restriction:
$$ \frac{\pi^2}{4} \leq \frac{2mV_0}{\hbar^2}L^2 \leq \frac{9 \pi^2}{4} \Rightarrow \frac{\pi^2 (\hbar c)^2}{8(m c^2)} \leq V_0 L^2 \leq \frac{9 \pi^2 (\hbar c)^2}{8(m c^2)}$$


By solving the problem, we obtain the solution $ \xi_1 $ and then the energy is calculated using the following formula:
$$ \xi_1 = k L \Rightarrow |E| = V_0 - \frac{(\hbar c)^2}{2(m c^2)L^2} \xi_1^2 $$

Our problem has two free parameters, $V_0$ and $L^2$. We solve the problem using different values of them in order to check which sets give energy values that align with the one measured from experiments ($ \pm 10\%$), which is:
$$ |E_b| = 2.224 MeV$$
Those sets of $V_0$ and $L^2$ must align with the previous restriction. In addition to this, $L$ is correlated with the radius of the atom, $ R \approx 3 fm$ , therefore its values should be in that order of magnitude.

```python
# Calculating the boundaries of V0*L2
lb = np.ceil(np.pi**2 * hbarc**2/(8*mc2))
ub = np.floor(9*np.pi**2 * hbarc**2/(8*mc2))
lb, ub
```

Therefore:
$$ 52  \leq V_0 L^2 \leq  460 $$ 

```python
# Setting the nu functions in order for graphical solving.
def n1(x):
    return -x/np.tan(x)
def n2(x, V0,L2):
    xmax = np.sqrt((2*mc2*V0*L2)/(hbarc**2))
    epsilon = 1e-10
    xcopy = x.copy()
    xcopy[xmax<xcopy] = xmax - epsilon
    return np.sqrt((2*mc2*V0*L2)/(hbarc**2) - xcopy**2)
def f(x,V0,L2):
    y = n2(x,V0,L2) - n1(x)
    return y
```

```python
# Plotting
ksi = np.linspace(0.01, 6, 2000)
xPlot1 = np.linspace(0.01,np.pi-1e-8,100)
xPlot2 = np.linspace(3.2, 6,100)
plt.plot(xPlot1,n1(xPlot1), color='k', label=r'$ \nu = - \xi cot( \xi)$')
plt.plot(xPlot2,n1(xPlot2), color='k')
plt.plot(ksi,n2(ksi,52,1), linewidth=0.8 ,linestyle='--', color='k', label=r'$ \nu = \sqrt{2mV_0 L^2 / \hbar^2 - \xi^2}$')#color='tab:blue', linestyle='--')
plt.text(0.6, 0.7, r'$V_0L^2$ = 52')
plt.plot(ksi,n2(ksi,150,1), linewidth=0.8 ,linestyle='--', color='k')#color='tab:orange', linestyle='--')
plt.text(1.3, 1.5, r'$V_0L^2$ = 150')
plt.plot(ksi,n2(ksi,300,1), linewidth=0.8 ,linestyle='--', color='k')#color='tab:orange', linestyle='--')
plt.text(1.5, 2.8, r'$V_0L^2$ = 300')
plt.plot(ksi,n2(ksi,460,1), linewidth=0.8 ,linestyle='--', color='k')#color='tab:blue', linestyle='--')
plt.text(3.3, 3.6, r'$V_0L^2$ = 460')
plt.ylim([0.1, 6])
plt.xlim([0, 6])
plt.xlabel(r'$\xi$')
plt.ylabel(r'$\nu ( \xi)$')
plt.title('Graphical Solution')
plt.legend(frameon=True, edgecolor='black', fontsize=10) 
plt.show()
```

As presented on the figure, the values 52 and 460 are indeed the boundaries for the problem to have a unique sollution. In the following part of the code, the problem is solved multiple times and when the calculated energy matches the desired one, we store the set on a dataframe.

```python
l_values = np.arange(1,4+0.01,0.01)
Eb = 2.224
El = Eb*0.9
Eu = Eb*1.1
```

```python
res = []
for l in l_values:
    L2 = l*l
    Vl = lb/L2
    Vu = ub/L2
    v_values = np.linspace(Vl,Vu,100)
    for v in v_values:
        sol = root(f, 1, args=(v,L2))
        ksi1 = sol['x']
        E = v - (hbarc*hbarc)/(2* mc2*L2)*ksi1**2
        if (El<=E and E<=Eu):
            res.append({'V0': v, 'L2': L2, '|E|': E[0]})
df = pd.DataFrame(res)
```

On the dataframe above, we have a table of all the accepted sets. But the problem is far from over. Although the bounding energy is inside the boundaries we have set, we still have not checked if the radius of the atom has a value that is accepted from the experimental data. Therefore, we need to calculate the radius of the atom for each of those sets of $V_0$ and $L^2$ and only keep the ones that are within the correct order of magnitude. 

The radius of the atom is obtained using:
$$ R = \sqrt{\langle r^2 \rangle}$$
where the mean value of $r^2$ is calculated from the following formula:
$$ \langle r^2 \rangle = 4 \pi \int_0^\infty \eta^2(r) r^2 dr$$
So, the first thing we need to calculate are the A and B parameters of the wavefunction. The first equation is obtained from the continuity of the wave function:
$$ A sin(k L) = B e^{-\gamma L}$$
The second equation is obtained from the normalization of the wavefunction, where:
$$ \int_0^\infty \eta^2(r)dr = 1 \Rightarrow A^2 \int_0^L sin^2(k r)dr + B^2 \int_L^\infty e^{-2\gamma r}dr = 1$$
Solving the system provides us the two equations for A and B:
$$ A = \left(\frac{L}{2} - \frac{sin(2 k L)}{4k} + \frac{sin^2(k L)}{2 \gamma} \right)^{-\frac{1}{2}} \\$$
$$ B = A sin(k L) e^{\gamma L}$$
After obtaining A and B, the calculation of $\langle r^2 \rangle$ is possible, as is the calculation of R. By doing some algebra, we derive the formula:
$$ \langle r^2 \rangle = 4 \pi \left( A^2 \int_0^L sin^2(kr)r^2dr + B^2 \int_L^\infty e^{-2\gamma r} r^2dr  \right)$$
(All the integrals are solved analytically using SymPy)

```python
def Radius(V,L,E):
    #Calculating k and gamma.
    k2 = (2*(mc2)*(V-E))/(hbarc*hbarc)
    k = np.sqrt(k2)
    g2 = (2*(mc2)*E)/(hbarc*hbarc)
    g = np.sqrt(g2)
    kL = k*L
    gL = g*L
    
    #Calculating A and B of the wavefunction.
    A = L/2 - np.sin(2*kL)/(4*k) + np.sin(kL)**2/(2*g)
    A = np.sqrt(1/A)
    B = A*np.sin(kL)*np.exp(g*L)
    
    a = L**3/6 - L**2*np.sin(2*L*k)/(4*k) - L*np.cos(2*L*k)/(4*k**2) + np.sin(2*L*k)/(8*k**3)
    b = np.exp(-2*g*L)*(1 + 2*g*L*(1+g*L))/(4*g**3)
    
    r2m = 4*np.pi*(A*A*a + B*B*b)
    R = np.sqrt(r2m)
    return [R,A,B,k,g]
```

After calculating the Radius for every set of $V_0$ and $L^2$ that gives optimal energy, we ONLY keep the sets whose radius is calculated to be between 1 and 10 fm.

```python
radii = []
for i in range(df.shape[0]-1):
    r = Radius(df['V0'][i],np.sqrt(df['L2'][i]),df['|E|'][i])
    if 1<r[0] and r[0]<10:
        radii.append({'V0': df['V0'][i], 'L2': df['L2'][i], '|E|': df['|E|'][i], 'R': r[0]})
df_final = pd.DataFrame(radii)
```

Finally, we are gonna provide a visualization of a wavefunction of this model. We chose the parameters of row number 3 and then calculated the wavefunction. After that the plotting takes place.

```python
def h(x, A, B, k, g, L):
    y = np.zeros_like(x)
    mask = (x > 0) & (x < L)
    y[mask] = A * np.sin(k * x[mask])
    y[~mask] = B * np.exp(-g * x[~mask])
    return y
```

```python
xt = np.linspace(0.001,20,400)
```

```python
_, At, Bt, kt, gt = Radius(df_final['V0'][3], df_final['L2'][3], df_final['|E|'][3])
Lt = np.sqrt(df_final['L2'][3])
Vt = df_final['V0'][3]
ht = h(xt,At,Bt,kt,gt,Lt)
```

```python
plt.plot(xt,ht,linewidth=1.2,color='k')
plt.text(18, 0.65, 
         r'$V_0 = {:.2f}$'.format(Vt) + '\n' + 
         r'$L^2 = {:.3f}$'.format(Lt),
         fontsize=12,
         ha='center',
         va='center',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
plt.title('Wavefunction')
plt.xlabel(r'$r$ [fm]')
plt.ylabel(r'$\eta (r)$')
plt.show()
```

A last verification of our normalization is checking the area under the curve of the squared wavefunction. According to our normalization it should be equal to one.

```python
dxt = xt[1]-xt[0] # dx
I = np.trapz(ht**2, dx=dxt) # Integral
print(f'The integral of the squared wavefunction is equal to {I:.4f}.')
```

The value of the integral proves that the normalization was indeed correct. The value is not exactly $1.0$, since the squared wavefunction was integrated from 0 to 20 fm, rather than 0 to $\infty$ which is implied by theory.
